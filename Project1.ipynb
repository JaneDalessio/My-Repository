> #### (6) Within your dataset, group the data based upon a variable and calculate 3 different aggregations. If your data are not suitable to grouping, just calculate the 3 aggregations for the whole dataset. What do these aggregations suggest about the data?

```python
numeric_columns = ["streams", "danceability_%", "energy_%"]
for col in numeric_columns:
    data[col] = pd.to_numeric(data[col], errors="coerce")
# I kept getting an error and concluded that some of the numbers were actually strings, so I ran this to make sure all the values are numeric

aggregations = data.groupby("key").agg({"streams": "mean", "danceability_%": "median",  "instrumentalness_%": "mean"}).reset_index()

aggregations
```

By aggregating the data by average streams, I found that among the 953 top-streamed songs on Spotify in 2023, those in the key of C# had the highest average streams. Next, I analyzed the median danceability percentage for songs grouped by key and discovered that songs in the key of B had the highest median danceability. Lastly, I examined the average instrumentalness percentage by key, revealing that songs in the key of F had the highest average instrumentalness, while those in A# had the lowest.


> #### (7)	Create two different types of visualizations with your dataset. These must be different kinds of visualizations (ex. scatterplot and histogram, not two histograms). Label them appropriately. Outline what they mean.

```python
plt.scatter(data['released_year'], data['speechiness_%'], alpha = 0.7, color = 'purple')
plt.xlabel(' Song Release Year ')
plt.ylabel('Amount of Spoken Words (%)')
plt.title('Correlation Between Release Year and Spoken Words Top Spotify Songs (2023)')
plt.show()
```

This scatter plot shows the relationship between the amount of spoken words in a song (speechiness) and its release year. From the graph, it seems that there is a weak positive correlation between the year a song was released and the amount of spoken words in that song, with more recent songs appearing to have higher speechiness scores. However, the relationship is not strong enough to draw a definitive conclusion or form a robust hypothesis.

```python
plt.hist(data['artist_count'], color = 'orange', bins=range(1, data['artist_count'].max() + 1))
plt.xlabel('Number of Artists per Song')
plt.ylabel('Frequency')
plt.title('Distribution of Artist Count in Top Spotify Songs from 2023')
plt.show()
```

This graph illustrates the distribution of the number of artists featured in the top songs of 2023. From the data, it is apparent that most of these popular songs were created by solo artists. However, there are also songs featuring up to seven artists. Overall, the trend shows that the majority of top songs this year were primarily performed by a single artist.
